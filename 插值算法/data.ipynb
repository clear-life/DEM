```python
# 构造一个数据生成器，参数包括数据的维度m，在一个范围内随机生成10W组数据
# 首先随机选择一个基准值b[1-1000]，然后基于这个基准值选择一个随机域度b范围内，然后在这个域度内随机生成指定维度的数据
# 再根据公式计算标签数据，做一个高斯噪声，方差为随机域度的0.05
# 然后将数据保存至文件m_u_re_data.csv

e = {}
e[1] = [0.0, 3.230346, 6.365597, 9.287162]
e[2] = [0.0, -3.511123, -2.150457, 0.765123]
e[3] = [0.0, -0.003001, -0.004572, 0.100655]
e[4] = [0.0, 4.431211, -39.770867, -8.100655]
e[10] = [0.0, 0.430211, 0.300124, 0.200563, 0.100278, 0.060042, 0.030086, 0.020019, 0.010045, 0.006258, -0.005583, 
         -0.005593, -0.005801, -0.006241, -0.008119, -0.016237, -0.019347, -0.020067, -0.020935, -0.021559, -0.023122, 
         -0.026291, -0.028590, -0.0292467, -0.031117, -0.033642, -0.036798, -0.037305, -0.037305, -0.037426, -0.037635, 
         -0.037702, -0.037833, -0.037902, -0.038032, -0.038126, -0.038243, -0.038333, -0.038428, -0.038509, -0.038611,
         -0.038708, -0.038791, -0.038875, -0.038963, -0.039041, -0.039132, -0.039220, -0.039307, -0.039392, -0.039485]
u = np.array([1,2,3,4])
alpha = 0.6
beta = 0.001
l = 10 #距离为10m
k = 10.0
raw = 100000 #10W

def gen_m_u_re_data(RE, m, u):
    # 创建一个指定维度的二维数组
    col = m+1
    data = np.zeros(shape=(raw, col), dtype=float)
    delta = []
    for i in range(raw):
        base = random.randint(30,1000)
        dot = random.randint(1,l)
        for j in range(col):
            if j == m:
                # 不同采样点的距离数组不一样
                if m == 4:
                    d = np.array([1,1,1,2]) * l
                    wu = []
                    for di in range(d.size):
                        floor = math.floor(d[di])
                        wi = alpha*d[di]**(-u*k) + (u/k) + e[k][floor-1] + (d[di]-floor)*(e[k][floor]-e[k][floor-1]) #加e项，小数的时候按比例折算
                        wu.append(wi)
                    w = np.array(wu/sum(wu)) # 计算权重
                    Zp = round(w.dot(data[i,:m].T),3)
                    delt = round(random.gauss(0, RE[u-1]),3)
                    data[i,j] = round(Zp + delt,1)
                    delta.append(Zp - data[i,j]) #残差
                elif m == 10:
                    d = np.array([math.sqrt(2),1,math.sqrt(2),2,1,1,2,math.sqrt(2),1,math.sqrt(2)]) * l
                    wu = []
                    for di in range(d.size):
                        floor = math.floor(d[di])
                        wi = alpha*d[di]**(-u*k) + (u/k) + e[k][floor-1] + (d[di]-floor)*(e[k][floor]-e[k][floor-1]) #加e项，小数的时候按比例折算
                        wu.append(wi)
                    w = np.array(wu/sum(wu)) # 计算权重
                    Zp = round(w.dot(data[i,:m].T),3)
                    delt = round(random.gauss(0, RE[u-1]),3)
                    data[i,j] = round(Zp + delt,1)
                    delta.append(Zp - data[i,j]) #残差
                elif m == 16:
                    # 通过下标计算两点的距离，选点的方式为固定的16个点：(-2,0),(-1,-2~2),(0,-2~2),(1,-2~2),(2,0)
                    loc = np.array([[-2,0],[-1,-2],[-1,-1],[-1,0],[-1,1],[-1,2],[0,-2],[0,-1],[0,1],[0,2],[1,-2],[1,-1],[1,0],[1,1],[1,2],[2,0]])
                    # 距离数组
                    d = []
                    for li in range(loc.shape[0]):
                        d.append(math.sqrt((loc[li][0])**2 + (loc[li][1])**2)*l)
                    d = np.array(d)
                    wu = []
                    for di in range(d.size):
                        floor = math.floor(d[di])
                        wi = alpha*d[di]**(-u*k) + (u/k) + e[k][floor-1] + (d[di]-floor)*(e[k][floor]-e[k][floor-1]) #加e项，小数的时候按比例折算
                        wu.append(wi)
                    w = np.array(wu/sum(wu)) # 计算权重
                    Zp = round(w.dot(data[i,:m].T),3)
                    delt = round(random.gauss(0, RE[u-1]),3)
                    data[i,j] = round(Zp + delt,1)
                    delta.append(Zp - data[i,j]) #残差
                elif m == 22:
                    loc = np.array([[-2,-2],[-2,-1],[-2,0],[-2,1],[-1,-2],[-1,-1],[-1,0],[-1,1],[-1,2],[0,-2],[0,-1],[0,1],[0,2],[1,-2],[1,-1],[1,0],[1,1],[1,2],[2,-1],[2,0],[2,1],[2,2]])
                    # 距离数组
                    d = []
                    for li in range(loc.shape[0]):
                        d.append(math.sqrt((loc[li][0])**2 + (loc[li][1])**2)*l)
                    d = np.array(d)
                    wu = []
                    for di in range(d.size):
                        floor = math.floor(d[di])
                        wi = alpha*d[di]**(-u*k) + (u/k) + e[k][floor-1] + (d[di]-floor)*(e[k][floor]-e[k][floor-1]) #加e项，小数的时候按比例折算
                        wu.append(wi)
                    w = np.array(wu/sum(wu)) # 计算权重
                    Zp = round(w.dot(data[i,:m].T),3)
                    delt = round(random.gauss(0, RE[u-1]),3)
                    data[i,j] = round(Zp + delt,1)
                    delta.append(Zp - data[i,j]) #残差
                elif m == 28:
                    loc = np.array([[-3,0],[-2,-2],[-2,-1],[-2,0],[-2,1],[-2,2],[-1,-2],[-1,-1],[-1,0],[-1,1],[-1,2],[0,-3],[0,-2],[0,-1],[0,1],[0,2],[0,3],[1,-2],[1,-1],[1,0],[1,1],[1,2],[2,-2],[2,-1],[2,0],[2,1],[2,2],[3,0]])
                    # 距离数组
                    d = []
                    for li in range(loc.shape[0]):
                        d.append(math.sqrt((loc[li][0])**2 + (loc[li][1])**2)*l)
                    d = np.array(d)
                    wu = []
                    for di in range(d.size):
                        floor = math.floor(d[di])
                        wi = alpha*d[di]**(-u*k) + (u/k) + e[k][floor-1] + (d[di]-floor)*(e[k][floor]-e[k][floor-1]) #加e项，小数的时候按比例折算
                        wu.append(wi)
                    w = np.array(wu/sum(wu)) # 计算权重
                    Zp = round(w.dot(data[i,:m].T),3)
                    delt = round(random.gauss(0, RE[u-1]),3)
                    data[i,j] = round(Zp + delt,1)
                    delta.append(Zp - data[i,j]) #残差
                elif m == 34:
                    loc = np.array([[-3,-2],[-3,-1],[-3,0],[-3,1],[-2,-2],[-2,-1],[-2,0],[-2,1],[-2,2],[-1,-2],[-1,-1],[-1,0],[-1,1],[-1,2],[0,-3],[0,-2],[0,-1],[0,1],[0,2],[0,3],[1,-2],[1,-1],[1,0],[1,1],[1,2],[2,-2],[2,-1],[2,0],[2,1],[2,2],[3,-1],[3,0],[3,1],[3,2]])
                    # 距离数组
                    d = []
                    for li in range(loc.shape[0]):
                        d.append(math.sqrt((loc[li][0])**2 + (loc[li][1])**2)*l)
                    d = np.array(d)
                    wu = []
                    for di in range(d.size):
                        floor = math.floor(d[di])
                        wi = alpha*d[di]**(-u*k) + (u/k) + e[k][floor-1] + (d[di]-floor)*(e[k][floor]-e[k][floor-1]) #加e项，小数的时候按比例折算
                        wu.append(wi)
                    w = np.array(wu/sum(wu)) # 计算权重
                    Zp = round(w.dot(data[i,:m].T),3)
                    delt = round(random.gauss(0, RE[u-1]),3)
                    data[i,j] = round(Zp + delt,1)
                    delta.append(Zp - data[i,j]) #残差
            else:
                # 随机生成值
                # 随机选择一个基准值b，然后基于这个基准值选择一个随机域度b范围内，然后在这个域度内随机生成指定维度的数据
                data[i,j] = round(random.uniform(base-dot,base+dot),1)

    #保存数据
    filename = "data/RE_m" + str(m) + "_u" + str(u) + "_data.csv"
    with open(filename,'w',newline='') as t:#numline是来控制空的行数的
        writer=csv.writer(t)#这一步是创建一个csv的写入器
        writer.writerows(data)#写入样本数据
    
    # 验证数据
    delta = np.array(delta)
    # print(RE[u-1], delta)
    Q = sum(delta**2) /raw
    p = sum(delta) / raw
    error = math.sqrt(sum(abs(delta - p)**2)/(raw-1))
    print("m=", m, ",u=", u, ",K=",k ,",Q=", Q,",Residual=",p,",Error=", error)

# 尝试生成所有数据文件
# load RE data
RE_m4 = np.array([0.00,0.031,0.14,0.24]) #0.02,0.04,0.14,0.24
RE_m10 = np.array([0.025,0.09,0.19,0.29]) #0.035,0.09,0.19,0.29
RE_m16 = np.array([0.05,0.15,0.25,0.35]) #0.05,0.15,0.25,0.35
RE_m22 = np.array([1.46,1.55,1.07,1.52])
RE_m28 = np.array([1.04,1.15,0.73,1.21])
RE_m34 = np.array([0.52,0.7,0.33,0.85])
uarray = np.array([1,2,3,4])

# RE_m4
for u in uarray:
    gen_m_u_re_data(RE_m4, 4, u)
# RE_m10
for u in uarray:
    gen_m_u_re_data(RE_m10, 10, u)
# RE_m16
for u in uarray:
    gen_m_u_re_data(RE_m16, 16, u)
# RE_m22
for u in uarray:
    gen_m_u_re_data(RE_m22, 22, u)
# RE_m28
for u in uarray:
    gen_m_u_re_data(RE_m28, 28, u)
# RE_m34
for u in uarray:
    gen_m_u_re_data(RE_m34, 34, u) 
```
